```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gross_rev = np.linspace(1000,2e5,1000)  # USD
x = gross_rev

# Calculating percentages of gross revenue costs for various licenses
houdini = np.ones(len(x))*269/x
unreal = np.zeros(len(x))
steam = 0.3
steam_publish_fee = 100/x
unreal[x>1e6] = 0.05
houdini[x>1e5] = 4500/x[x > 1e5]

total_cost_of_gross = houdini +  steam + steam_publish_fee + unreal
y = total_cost_of_gross
x = gross_rev * 6.62
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
#plt.loglog(x,y)
plt.grid(True)
plt.ticklabel_format(style="sci")
plt.xticks(np.linspace(x.min(),x.max(), 17))
plt.yticks(np.linspace(y.min(),y.max(), 17))

plt.ylabel(r"Cost of licenses [% Gross Revenue]", fontsize="16")
plt.xlabel(r"Gross Revenue [DKK]", fontsize="16")
```

```python
rev = 1e6 # dkk
price_pr_game = 15 # eur
rev_pr_game = price_pr_game * 7.5 # dkk

fee = rev * y[ np.abs(rev - x).argmin() ] # find nearest neighbour to the revenue

print(f"Amount of games sold before we have revenue of {rev} DKK")
print(rev/rev_pr_game, f"At the price of {price_pr_game} euro pr game", "\n")

print("We pay", fee , "DKK")
print("Leaving us with profit of", rev - fee, "DKK", "\n")

print("This is given a period of 1 year, this is the least efficient amount to sell.")
print("as you pay the highest percentage of revenue to licenses.")
```
